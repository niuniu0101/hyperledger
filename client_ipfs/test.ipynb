```python
def split_cids_to_servers(input_file_path):
        # 初始化服务器文件的写入器
        server_writers = {}
        
        try:
            # 读取输入文件
            with open(input_file_path, 'r') as input_file:
                for line in input_file:
                    # 去除行首尾的空白字符并分割
                    parts = line.strip().split()
                    if len(parts) != 2:
                        print(f"跳过无效行: {line.strip()}")
                        continue
                    
                    cid, server_num = parts
                    server_file = f"server{server_num}.txt"
                    
                    # 如果该服务器文件的写入器不存在，则创建
                    if server_file not in server_writers:
                        server_writers[server_file] = open(server_file, 'w')
                    
                    # 将CID写入对应的服务器文件
                    server_writers[server_file].write(cid + '\n')
                    print(f"已将 {cid} 写入 {server_file}")
        
        except FileNotFoundError:
            print(f"错误: 找不到文件 {input_file_path}")
        except Exception as e:
            print(f"处理文件时发生错误: {str(e)}")
        finally:
            # 关闭所有打开的文件
            for writer in server_writers.values():
                writer.close()

if __name__ == "__main__":
    # 输入文件路径
    input_file = "/root/go/src/github.com/hyperledger/client_ipfs/file_to_ring.txt"
    # 执行分割操作
    split_cids_to_servers(input_file)
    print("操作完成")
    
```
